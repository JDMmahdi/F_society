```python
import matplotlib.pyplot as plt
```

```python
temprature_xval = [0,20,40,60,80,100]
solub_yval = [25.5,36.2,48,60,70,83]
plt.title("Solubility Of CuSO4 in Water", color = 'red')
plt.xlabel("TEMPRATURE")
plt.ylabel("SOLUBILITY")
plt.scatter(temprature_xval, solub_yval, cmap="viridis")

plt.colorbar()
```

```python
temp_str = input("Please enter the temperature values (comma-separated): ")
temp_list = temp_str.split(",")
temprature_xvals = []
for temp in temp_list:
    while not temp.isdigit():
        temp = input("Invalid input. Please enter a valid integer value for the temperature: ")
    temprature_xvals.append(float(temp))


solub_str = input("Please enter the solubility values (comma-separated): ")
solub_list = solub_str.split(",")
solub_yvals = []
for solub in solub_list:
    while not solub.isdigit():
        solub = input("Invalid input. Please enter a valid integer value for the solubility: ")
    solub_yvals.append(float(solub))

comp_name = input("Please enter the name of compound or element: ")
plt.title("Solubility Of %s In Water" %(comp_name), color='hotpink')
plt.scatter(temprature_xvals, solub_yvals, cmap = "viridis")
plt.colorbar()
plt.grid()
```
