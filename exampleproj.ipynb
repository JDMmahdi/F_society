```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_label_t = input("Please enter the period of time you considered (in months): ")
y_label_sales_val = input("Please enter VALUE of SALES you considered: ")

fig_sales = plt.figure(figsize=(8,8))
plt.title("SALES")
months = range(1, int(x_label_t) + 1)
plt.subplot(2,2,1)
plt.plot(months, [y_label_sales_val] * len(months))

cost = int(input("Please enter the amount of costs: "))
y_label_income = int(y_label_sales_val) - cost

fig_income = plt.figure(figsize=(8,8))
plt.title("INCOME")
plt.subplot(2,2,3)
plt.plot(months, [y_label_income] * len(months))

plt.show()
```
