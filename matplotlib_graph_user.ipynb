```python
import matplotlib.pyplot as plt
import re
```

```python
print("keep in mind that the x and y values should end at a same value ;)")
x = input("please input the x values: ")
x = x.replace(",", "")
y = input("please input the y values: ")
y = y.replace(",", "")

x = [int(i) for i in x]
y = [int(i) for i in y]

plt.plot(x,y)

plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
x = []
y = []

x_count = int(input("Please enter the NUMBER of X values that you want to have: "))
x_vals = input("please enter the X values for the X axis with ',' between them: ")
x_vals = x_vals.split(",")
for val in x_vals:
    x.append(int(val))


y_count = input("Please enter the NUMBER of Y values that you want to have: ")
y_vals = input("Please enter the Y values for the Y axis with ',' between them: ")
y_vals = y_vals.split(",")
for val in y_vals:
    y.append(int(val))

plt.plot(x,y)
plt.show()
```
